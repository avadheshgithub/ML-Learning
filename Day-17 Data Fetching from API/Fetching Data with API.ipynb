```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
pd.DataFrame(response.json()['results'])
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df.head()
```

```python
df = pd.DataFrame()
```

```python
df
```

```python

for i in range(1, 429):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')
    
    # Extract data from JSON response
    temp_df = pd.DataFrame(response.json().get('results', []), 
                           columns=['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count'])

    # Use pd.concat instead of append
    df = pd.concat([df, temp_df], ignore_index=True)

# Display the first few rows
print(df.head())
```
