```python
import requests
from bs4 import BeautifulSoup

# Define headers to mimic a real browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

# Replace with the actual portfolio URL
url = "https://avadheshgithub.github.io/My_Portfolio/"

# Fetch the webpage
response = requests.get(url, headers=headers)

# Check if request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'lxml')  # Parse HTML content
    
    # Extract different sections safely
    def get_text_or_default(tag):
        return tag.text.strip() if tag else "Not Found"

    name = get_text_or_default(soup.find("h1"))
    role = get_text_or_default(soup.find("h2"))
    about_me = get_text_or_default(soup.find("section", id="about"))
    stats = get_text_or_default(soup.find("section", id="stats"))
    skills = get_text_or_default(soup.find("section", id="skills"))
    timeline = get_text_or_default(soup.find("section", id="timeline"))
    portfolio = get_text_or_default(soup.find("section", id="portfolio"))

    # Organizing and displaying extracted data
    organized_data = f"""
    # {name}
    ## {role}

    ## About Me
    {about_me}

    ## My Stats
    {stats}

    ## My Skills
    {skills}

    ## My Timeline
    {timeline}

    ## My Portfolio
    {portfolio}
    """

    print(organized_data)  # Print extracted content

else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```
