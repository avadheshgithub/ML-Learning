## EDA (Univariate Analysis)

```python

## Univariate Analysis - at a time only analysis of one variable (column)

## Bivariate Analysis - at a time  analysis of two variable (column)

## multivarient Analysis - more than two varibales at a time
```

```python
Data Type (column)

    - Numerical
    - catogorical
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

## 1. Categorical Data

### a. Countplot

```python
sns.countplot(df['Embarked'])
```

```python

df['Survived'].value_counts().plot(kind='bar')
```

### b. PieChart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

## 2. Numerical Data

### a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

### b. Distplot

```python
sns.distplot(df['Age'])
```

### c. Boxplot

```python
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['Age'])
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
df['Age'].mean()
```

```python
df['Age'].skew()

#      0 --> perfect
#      + --> + skewed
#      - --> - skewed
```
