## 1. OneHotEncoding using Pandas

```python
## Get dummies function

pd.get_dummies(df,columns=['fuel','owner'],dtype=int)
```

##  K-1 OneHotEncoding using pandas

```python
## remove first column

pd.get_dummies(df,columns=['fuel','owner'],drop_first=True,dtype=int)
```

## 2. OneHotEncoding using Sklearn

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']]).toarray()
```

```python
X_test_new = ohe.transform(X_test[['fuel','owner']]).toarray()
```

## 3. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
repl
```

```python
## when we have multiple column then --> make one for all into one column
pd.get_dummies(df['brand'].replace(repl, 'uncommon'),dtype=int).sample(5)
```
