## Making our own Linear Regression Class


```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('placement.csv')
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
class MeraLR:
    
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
        
    def fit(self,X_train,y_train):
        num = 0
        den = 0
        
        for i in range(X_train.shape[0]):
            num = num + ((X_train[i]-X_train.mean())*(y_train[i] - y_train.mean()))
            den = den + ((X_train[i]-X_train.mean())*(X_train[i] - X_train.mean()))

        self.m = num/den
        self.b = y_train.mean() - (self.m * X_train.mean())
        print(self.m)
        print(self.b)
        
    def predict(self,X_test):
        return self.m * X_test + self.b
    
```

```python
lr = MeraLR()
```

```python
lr.fit(X_train,y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a 2D function (e.g. a "loss surface")
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) + 0.1 * x + 0.2 * y

# Create meshgrid
x = np.linspace(-6, 6, 60)
y = np.linspace(-6, 6, 60)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Compute gradient
dx, dy = np.gradient(Z)

# 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='jet', alpha=0.8)

# Pick a point (like in the image) to visualize arrows
ix, iy = 30, 30  # center point index
x0, y0, z0 = X[ix, iy], Y[ix, iy], Z[ix, iy]
grad_x = dx[ix, iy]
grad_y = dy[ix, iy]

# Normalize the gradient vector
norm = np.sqrt(grad_x**2 + grad_y**2)
grad_x /= norm
grad_y /= norm

# Gradient vector (steepest ascent)
ax.quiver(x0, y0, z0, grad_x, grad_y, 1, color='r', linewidth=2, label='Gradient (Ascent)')
# Negative gradient vector (steepest descent)
ax.quiver(x0, y0, z0, -grad_x, -grad_y, -1, color='g', linewidth=2, label='-Gradient (Descent)')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Loss')
ax.set_title("Gradient and Descent Direction on Loss Surface")
ax.legend()

plt.show()
```
